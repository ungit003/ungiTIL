### 화면 예시
- 상품 목록 페이지
    
    - Card 형태로 상품 정보를 출력
    
    - 버튼 종류
        - 상세 페이지로 이동
        
        - 장바구니 추가



- 장바구니에 담은 상품 목록은 Local Storage 에 저장합니다.

- 상품 상세 페이지
    - 상품의 상세 정보를 출력합니다.
    
    - 버튼 종류
        - 장바구니에 추가

- 장바구니 페이지
    - 현재까지 장바구닝 담은 상품 목록을 출력합니다.
    
    - 버튼 목록
        - 상세 페이지로 이동
        
        - 장바구니에서 제거
