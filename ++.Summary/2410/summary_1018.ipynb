- 웹 개발자에게 "회원제 게시판"은 의미가 큽니다.
- 거의 모든 웹 애플리케이션에서 사용되는 `핵심 기술들을 포함`하고 있습니다.
    1. CRUD
    2. 회원 가입, 로그인
