# Data Track

## Tree

- 비선형 구조
    - 선형 구조
        - 데이터 사이에 순서 존재
        - 1:1 관계로 연결
        - stack, que, list, 연결리스트
    - 순차적 x
    - tree, graph

- 1:N, 원소간 계층 관계, 상위 원소 -> 하위 원소로 내려가면서 확장되는 나무모양 구조(위아래를 뒤집어서 보면)

- 한 개 이상의 노드로 이루어진 유한 집합
    - 최상위 노드 : 루트(root)
    - 나머지 노드 : 분리 집합이 될 수 있다 = 짜르면 걔네가 다시 트리가 된다. = 트리는 트리를 모아서 만들었다.

- 노드
    - 트리의 원소

- 간선
    - 노드와 노드를 연결하는 선
    - 부모, 자식 노드를 연결

- 루트 노드
    - 트리의 시작 노드, 최상위 노드

- 형제 노드
    - 같은 부모 노드의 자식 노드

- 조상 노드
    - 루트 노드까지 가는 간선을 따라 존재하는 모든 노드

- 서브 트리
    - 부모 노드와 연결된 간선을 끊었을 때 생성되는 트리

- 자손 노드
    - 서브 트리에 있는 하위 레벨의 노드들

- 차수(degree) = 자(식)수
    - 노드에 연결된 자식 노드 수
    - 트리의 차수
        - 트리에 있는 노드의 차수 중에 제일 큰 값
        - 단말 노드
            차수가 0인 노드 = 자식x = leaf

- 레벨
    루트 ~ 노드 거리
    루트 레벨 = 0, 자식은 내려갈 수록 + 1

- 높이
    - 노드에서 가장 먼 리프 노드까지의 간선 수
    - 트리의 높이는? 의 느낌 (0부터 시작)

## Binary Tree
- 차수가 2인 트리(자식이 두개만 있음)
- 각 노드가 자식 노드를 최대한 2개 까지만 가질 수 있음 = 모든 노드는 최대 2개의 서브 트리를 가짐
- 높이 i = 최대 노드 2^i 최소 노드 i + 1

- 종류
    - 포화 이진 트리
    - 완전 이진 트리
    - 편향 이진 트리

- 리스트를 이용한 이진 트리의 표현
    - 편향 트리로 짜면 빈칸이 너무 많아
    

## Traversal

- 순회 : 트리 노드를 체계적으로 방문하는 것
- 방법
    - 전위 : 루트 왼 오 : 디렉토리 방문
    - 중위 : 왼 루트 오 : 정렬된 순서로 출력 (BST)
    - 후위 : 왼 오 루트 : 디렉토리 총 크기 구하기 = 제일 안쪽의 친구들 방문

- 수식 트리
    - 후위가 더 좋은가?

## Heap

우선순위 큐 : 들어온 애들한테 순번을 매기는 것. 최단거리, 문자열 압축


- 완전 이진 트리, 키값이 가장 큰 노드 & 키 값이 가장 작은 노드를 찾기 위해 만든 구조
- 최대 힙
    - 키값 가장 큰 거 찾는 거
    - 부모 노드의 키값 > 자식 노드의 키값
    - 루트 노드 : 키값이 가장 큰 노드
- 최소 힙
    - 키값 가장 작은거 찾기
    - 부모 노드 키값 < 자식 노드 키값
    - 루트 노드 : 키값 가장 작으 노드

- 힙 연산
    - 삽입
        - 무조건 마지막 자리(리프노드)에 빈자리를 만들어서 넣어야함(완전 이진 트리)
        - 부모보다 크면(작으면) 자리를 바꿔야함(루트까지) 

    - 삭제
        - 루트 노드의 원소만 삭제 가능(가장 크거나 작거나)
        - 우선순위 큐에 좋음. 힙구조의 특성
        - 루트 지우고 마지막 노드를 루트에 놓음
        - 삽입의 과정 반복. 루트에 왔는데 자식이 둘 다 크면 더 큰 애랑 바꾸기.
