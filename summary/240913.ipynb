## Client & Server

웹의 동작 방식



### Client
클라이언트 = `서비스`를 `요청`하는 주체
- -> 웹 사용자의 인터넷이 연결된 장치, 웹 브라우저

### Server
서버 =  클라이언트의 요청에 `응답(페이지, 로직, 파일 ~~)`하는 주체
- -> 웹 페이지, 앱을 저장하는 컴퓨터(서버 컴퓨터)

우리가 웹 페이지를 보는 과정.
1. 웹 브라우저(클라이언트)에서 `google.com` 입력, 엔터
2. 웹브라우저가 전세계 어딘가에 있는 구글 컴퓨터(서버)에 `메인 홈페이지.html`을 request
3. 요청 받은 구글 컴퓨터 : `메인 홈페이지.html` 파일을 찾아서 response
4. 웹브라우저 : 전달받은 `메인 홈페이지.html` 파일을 사람이 볼 수 있도록 해석. 사용자가 볼 수 있게 해줌

## Frontend & Backend

- Frontend (프론트엔드)
    - 사용자 인터페이스(UI) 구성. 사용자가 애플리케이션과 상호작용할 수 있도록 함

- `Backend (백엔드)`

    - -> 서버 언어(python, java ~~), 백엔드 프레임워크, 데이터베이스, API, 보안 등

- 이번에 공부할 부분

## `가상 환경(중요)`


Python 애플리케이션과 그에 따른 패키지들을 격리하여 관리할 수 있는 `독립적인 실행 환경`
- 기존의 환경 : pip install ~~ = 전역 global

why?
- case 1. 
    - 한 개발자가 2개의 프로젝트를 진행해야 함
    - 각각 A, B 패키지가 필요할 때, 한 번에 하나씩만 컴퓨터에 설치할 수 있음
    - 이럴 때 독립적인 개발 환경을 위해 가상 환경을 사용

- case 2.
    - 이번에도 2개의 프로젝트를 진행함
    - water, fire 패키지를 사용하는데, 두개를 동시에 설치하면 충돌이 발생.
    - 충돌을 피하기 위해서 각각 독립적인 개발 필요



### 가상환경 만들기

1. 가상 환경 venv 생성

```py
$ python -m venv venv
```
-> `venv 폴더 생성`

python -m venv : 가상환경 만들겠다

venv(뒤) : 가상환경 이름 -> 왜 이름이 이거임? 관례적으로 쓰는 것
**venv 폴더 안의 내용 절대 수정 x**

#### 가상 환경은 `on/off` 의 개념 in/out 아님.

2. 가상 환경 활성화
```py
$ source venv/Scripts/activate
```
-> `(venv) 출력`
-> (venv) 표시가 있다면. 가상 환경이 켜진 상태

venv/Scripts/activate : 경로
가상은 경로에 의해서 결정됨
-> 같은 venv 파일을 만들 수 있음 -> 그럼 구분 어떻게 함?
-> `경로`로 하면됨
-> 복잡해지니까 프로젝트와 함께(동일한 위치에) venv를 만들면 편하다.

3. 환경에 설치된 패키지 목록 확인
```py
$ pip list
```

가상 환경 활성화 하면 기존에 설치했던 pip들은 보이지 않게됨.

4. 설치된 패키지 목록 생성
```py
$ pip freeze > requirement.txt
```

pip freeze : pip 목록을 출력 하겠다.

`> requirements.txt` : 출력한 목록을 텍스트 파일에 적겠다.

이걸 왜 만드는가?
- case
    - A, B 협업
    - A가 먼저 가상 환경 만들고 github에 push
    - B가 clone을 해도 실행이 안됨
    - 프로젝트 안에 venv가 없음. 어떤 가상환경인지 알 수가 없음
    - -> `가상환경에 대한 정보 = 패키지 목록` 이 필요

venv를 채로 push하면 되는게 아닌가?
- pip를 몇개 설치하다보면 용량이 엄청나게 커짐.
- -> 설치한 목록만 알려주면 용량 절약 가능.
- 설치 코드
```py
$ pip install -r requirements.txt
```

+의존성 패키지
- 패키지를 설치하면 단순히 하나만 설치 되는 것이 아님
- 연관된 패키지가 전부 설치됨
-> 한 패키키지가 다른 패키지의 기능과 코드를 사용하기 때문에 그 패키지가 존재해야만 제대로 작동
-> 사용하려는 패키지가 설치되지 않았거나, 버전이 호환되지 않으면 오류발생, 이상한 작동함

5. 가상 환경 끄기

```py
$ deactivate
```
그냥 터미널을 꺼도 꺼지긴 한데 혹시나~ 하는 마음에

요약.
1. 가상환경은 on/off.임 (in/out x)
    - 활성화 하면 현재 터미널에만 영향을 줌
    - 새 창 열면 다시 활성화 해야함
2. `방`에 들어간다고 생각하지말고 `도구 세트` 라고 생각할 것.
3. 프로젝트마다 별도로 설정할 것
4. 프로젝트와 동일한 위치에 venv 폴더를 만들 것
5. 가상환경 폴더는 gitignore 에 들어있음. -> 레포지토리에 안올라감
    - 저장소 크기 줄이고 효율적인 협업, 배포를 위해서 requirements.txt파일만 공유

venv, requirements는 관례적인 것이니 수정하지 말 것.

-> 가상환경 & 의존선 패키지 관리를 잘 하자.

## Django Design Pattern

디자인 패턴
- 소프트웨어 설계에서 발생하는 문제를 해결하기 위한 일반적인 해결책
- -> 공통적인 문제를 해결하는 데 쓰이느 `형식화 된 관행`

-> `애플리케이션의 구조는 이렇게 구성하자` 라는 관행(약속)

MVC 디자인 패턴
- Model, View, Controller
- 애플리케이션을 구조화하는 대표적인 패턴
- 데이터(M), 사용자 인터페이스(V), 비즈니스 로직(C) 로 분리

MTV 디자인 패턴
- Model, Template(View의 역할), View(Controller의 역할)
- Django가 애플리케이션을 구조화하는 패턴
- 기존 MVC 패턴과 동일. 단순히 명칭을 다르게 함(파이썬에서는 ARR를 LIST로 쓰는 것처럼)
