### DTL Syntax
1. Variable
2. Filters
3. Tags
4. Comments

#### 1. Variable
- render 함수의 `세번째` 인자, 딕셔너리만 가능.
- {{key}}  
- 딕셔너리, 리스트 등 이중구조라면 : {{key.key}}



#### 2. Filters
- 표시할 변수를 `수정`할 때 사용 (변수 + `|` + 필터)
- chained(연결) 가능. 일부 필터는 인자를 받음
- 약 60개의 bulit-in template filters 제공(외우기 x 검색해서 사용)
- {{variable|filter}}  {{name|truncatewords:30}}



#### 3. Tags
- 반복, 논리를 수행하여 제어 흐름을 만듦
- 일부 태그는 시작과 종료 태그가 필요
- 약 24개의 bulit-in template filters 제공
- {%tag%}  {%if%} {%endif%}

### action & method
- form 핵심 속성 2가지
- 데이터를 `어디(action)`로 어떤 `방식(method)`으로 요청할건지?

- action
    - 입력 데이터가 전송될 URL을 지정 (목적지)
    - 만약 이 속성을 지정하지 않으면, 데이터는 현재form이 있는 페이지의 URL로 보내짐

- method
    - 데이터를 어떤 방식으로 보낼 것인지 정의
    - 데이터의 HTTP request method(GET, POST)를 지정
  

### 'input' element
- 사용자가 데이터를 입력 받을 수 있는 요소
- tpye 속성 값에 따라 다양한 유형의 입력 데이터를 받음
-> 핵심 속성 : `name`


#### 'name' attribute
- input의 핵심 속성
- 사용자가 입력한 데이터에 붙는 이름(key)
- 데이터를 제출했을 때 서버는 name 속성에 설정된 값을 통해서만 사용자가 입력한 데이터에 접근할 수 있음

### Query String Parameters
- 사용자의 입력 데이터를 URL 주소에 파라미터를 통해 서버로 보내는 방법
- 문자열을 앰퍼샌드(&)로 연결된 key=value 쌍으로 구성되며, 기본 URL과는 물음표(?)로 구분됨
- 예시
    - http://host:port/path?key=value&key=value
